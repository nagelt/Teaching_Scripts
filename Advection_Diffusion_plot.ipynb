# Graphical illustration of advection and diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
import plot_functions.plot_settings
```

```python
piecewise_function = lambda x, b: np.where((x >= b - 0.5) & (x <= b + 0.5), 1, 0)
```

```python
def smooth_pulse(x, b, width, smoothness):
    return 0.5 * (np.tanh((x - (b - width / 2)) / smoothness) - np.tanh((x - (b + width / 2)) / smoothness))
```

```python
x = np.linspace(0,10,400)
```

```python
v = 1.2
fig, ax = plt.subplots(figsize=(8,6))
for t in range(3):
    b = 5 + t * v
    alpha_value = (t + 1) / 3  # Increasing alpha values: 1/3, 2/3, 3/3
    ax.plot(x, piecewise_function(x, b), label=f'$t_{{{t}}}$', color='blue', alpha=alpha_value)

arrow = ax.arrow(0, 0.5, 2, 0, head_width=0.05, head_length=0.2, fc='black', ec='black')

ax.text(1, 0.55, r'$\vec{v}$', fontsize=12, ha='center')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
fig.suptitle(r'Advection')
ax.legend()
fig.savefig('advection.pdf')
```

```python
# Plot the function
fig, ax = plt.subplots(figsize=(8,6))
for t in range(3):
    smoothness = 0.001*(t+1)+0.15*t
    alpha_value = (t + 1) / 3  # Increasing alpha values: 1/3, 2/3, 3/3
    plt.plot(x, smooth_pulse(x, 5, 1, smoothness), label=f'$t_{{{t}}}$', color='blue', alpha=alpha_value)

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
fig.suptitle(r'Diffusion')
ax.legend()
fig.savefig('diffusion.pdf')
```
