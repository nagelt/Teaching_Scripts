## Beispiele

Wir schauen uns die Folgenden Beispiele an:

* Einachsiger Druck: $\sigma_1 > 0$, $\sigma_2 = \sigma_3 = 0$
* Einachsiger Zug: $\sigma_3 < 0$, $\sigma_1 = \sigma_2 = 0$
* Einfacher Schub: $\tau_{xy} \neq 0$, also $\sigma_1 > 0$, $\sigma_2 = 0$, $\sigma_3 = -\sigma_1$
* "Triaxialer" Druckversuch (unecht, da $\sigma_2 = \sigma_3$): $\sigma_1 > 0$, $\sigma_2 = \sigma_3 > 0$
