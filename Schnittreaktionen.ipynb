## Ebene Spannungszustände

Sei $\sigma_1 \geq \sigma_2 \geq \sigma_3$ und unsere Betrachtungen beschränkt auf die $1-3$ Ebene. Das Beispiel ist also analog zum obigen, aber mit einer zusätzlichen horizontal wirkenden Spannung auf die Probe (wie im Triaxialversuch). Ein Spannungszustand in einer unter dem Winkel $\alpha \in [0°,90°]$ geneigten Ebene ergibt sich dann über analoge Gleichgewichtsbetrachtungen aus den Hauptspannungen zu:
\begin{align*}
    \sigma_\alpha &= \underbrace{\frac{\sigma_1 + \sigma_3}{2}}_{\displaystyle \sigma_\text{M}} + \frac{\sigma_1 - \sigma_3}{2} \cos (2\alpha)
    \\
    \tau_\alpha &= \underbrace{\frac{\sigma_1 - \sigma_3}{2}}_{\displaystyle \tau_\text{max}} \sin (2\alpha)
\end{align*}

Durch Quadrieren und Aufaddieren lässt sich daraus eine Kreisgleichung wie folgt ableiten:

\begin{align*}
    \tau_\text{max}^2 &= \left( \sigma_\alpha - \sigma_\text{M} \right)^2 + \tau_\alpha^2
\end{align*}

Daraus ergeben sich die oben dargestellten Spannungskreise.
