## Linear aufgebrachte und anschließend konstante Flächenlast mit unendlicher Ausdehnung

Die Flächenlast $q_0$ wird wie folgt aufgebracht:

\begin{align*}
    q = \begin{cases}
        q_0 \frac{T}{T_\text{B}} & T \leq T_\text{B}
        \\
        q_0 & T > T_\text{B}
    \end{cases}
\end{align*}

Der Strömungsdruck stellt sich wie folgt ein

\begin{align*}
    \Delta u = \begin{cases}
         \frac{q_0}{T_\text{B}} \frac{16}{\pi^3} \sum \limits_{i=0}^\infty \left[ \frac{(-1)^i}{(2i+1)^3} \cos \left[ (2i+1) \frac{\pi}{2} \frac{z}{H} \right] \left( 1 - \exp \left( -(2i+1)^2 \frac{\pi^2}{4} T_\text{V}\right) \right)\right] & T \leq T_\text{B}
        \\
        \frac{q_0}{T_\text{B}} \frac{16}{\pi^3} \sum \limits_{i=0}^\infty \left[ \frac{(-1)^i}{(2i+1)^3} \cos \left[ (2i+1) \frac{\pi}{2} \frac{z}{H} \right] \left( \exp \left( -(2i+1)^2 \frac{\pi^2}{4} (T_\text{V} - T_\text{B}) \right) - \exp \left( -(2i+1)^2 \frac{\pi^2}{4} T_\text{V}\right) \right)\right] & T > T_\text{B}
    \end{cases}
\end{align*}
